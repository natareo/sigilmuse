```python
# 1.  Load images and set weights.
import numpy as np
from imageio import imread
```

```python
def crop(image,xborder,yborder):
    return image[xborder:image.shape[0]-xborder,yborder:image.shape[1]-yborder]
```

```python
from glob import glob
sigilfiles = glob('../sigilbank/png/256/*.png')

sigils  = {}
for sigilfile in sigilfiles:
    sigil = imread(sigilfile)[:,:,0:3]
    # Crop sigilfile.
    sigil = crop(sigil,76,76)
    # Add to dictionary.
    sigils[sigilfile[-7:-4]] = sigil.copy()
```

```python
import matplotlib.pyplot as plt
plt.imshow(sigils['fyl'])
```

```python
from cv2 import resize,INTER_CUBIC

weights = {}
for sigil in sigils:
    weight = resize(sigils[sigil],dsize=(4,4),interpolation=INTER_CUBIC)
    weights[sigil] = weight
```

```python
weights
```

```python
plt.imshow(weights['reb'])
```

```python
def distance(arr1,arr2):
    return np.sum(np.abs(arr1-arr2))

def match(target,sources):
    best_match = 'bal'
    best_distance = 1e8
    for key in sources:
        d = distance(sources[key],target)
        if d < best_distance:
            best_match = key
            best_distance = d
    return best_match
```

```python
# Scan through images in blocks of 4×4 pixels.
# Match each one to a weight.

source = imread('mars.png')
print(source.shape)
for row in range(source.shape[0]):
    for col in range(source.shape[1]):
        if source[row,col,:][-1] == 0:  source[row,col] = (0,0,0,0)
source = source[:,:,0:3]

if (source.shape[0]%4 != 0) or (source.shape[1]%4 != 0): assert False,'Image must be divisible by four.'

result = np.array([['' for y in range(0,source.shape[1],4)] for x in range(0,source.shape[0],4)],dtype=object)

from itertools import product
for xindex,yindex in product(range(0,source.shape[0],4),range(0,source.shape[1],4)):
    image_weight = source[xindex:xindex+4,yindex:yindex+4]
    result[xindex//4,yindex//4] = match(image_weight,weights)
```

```python
result
```

```python
# Convert back to an image composed of sigils.
from glob import glob
sigilfiles = glob('../sigilbank/png/32/*.png')

sigils  = {}
for sigilfile in sigilfiles:
    sigil = imread(sigilfile)[:,:,0:3]
    # Crop sigilfile.
    sigil = crop(sigil,10,10)
    # Add to dictionary.
    sigils[sigilfile[-7:-4]] = sigil.copy()

target = np.zeros((result.shape[0]*sigils['bal'].shape[0],result.shape[1]*sigils['bal'].shape[1],3))
xint = sigils['bal'].shape[0]
yint = sigils['bal'].shape[1]
for r,row in enumerate(result):
    for c,col in enumerate(row):
        target[r*xint:r*xint+xint,c*yint:c*yint+yint] = sigils[result[r,c]]
```
